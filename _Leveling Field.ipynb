```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Sample numeric data
data = {
    'Age': [22, 30, 45, 50, 28],
    'Salary': [25000, 60000, 120000, 150000, 50000]
}

df = pd.DataFrame(data)

std_scaler = StandardScaler()
df_standardized = pd.DataFrame(
    std_scaler.fit_transform(df),
    columns=df.columns
)


minmax_scaler = MinMaxScaler()
df_normalized = pd.DataFrame(
    minmax_scaler.fit_transform(df),
    columns=df.columns
)

print("Original Data:\n", df)
print("\nStandardized Data:\n", df_standardized)
print("\nNormalized Data:\n", df_normalized)
```

```python
plt.figure()
plt.hist(df['Salary'], bins=5)
plt.title("Salary Before Scaling")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()

plt.figure()
plt.hist(df_standardized['Salary'], bins=5)
plt.title("Salary After Standardization")
plt.xlabel("Scaled Salary")
plt.ylabel("Frequency")
plt.show()
```
