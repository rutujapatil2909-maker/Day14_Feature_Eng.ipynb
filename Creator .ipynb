```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

# Feature with non-linear relationship
X = np.arange(-10, 10, 1).reshape(-1, 1)
y = X.flatten()**2 + np.random.normal(0, 10, size=len(X))

# Convert to DataFrame
df = pd.DataFrame({'X': X.flatten(), 'y': y})

# Visualize
plt.figure()
plt.scatter(df['X'], df['y'])
plt.xlabel("X")
plt.ylabel("y")
plt.title("Non-linear Relationship between X and y")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Linear Regression on original feature
lr = LinearRegression()
lr.fit(X, y)

y_pred_linear = lr.predict(X)
r2_linear = r2_score(y, y_pred_linear)

print("R² Score (Linear Features):", r2_linear)
```
